```python
import os
import csv, glob

Dir = r"C:/Users/Stu/Desktop/SMU_Bootcamp/SMUDAL201905DATA5/02-Homework/20-Tableau/Instructions/Data"
Avg_Dir = r"C:/Users/Stu/Desktop/SMU_Bootcamp/SMUDAL201905DATA5/02-Homework/20-Tableau/Instructions/Data"

csv_file_list = glob.glob(os.path.join(Dir, '*.csv')) # returns the file list
print (csv_file_list)

with open(os.path.join(Avg_Dir, 'Output.csv'), 'w', newline='') as f:
    wf = csv.writer(f, lineterminator='\n')

    for files in csv_file_list:
        with open(files, 'r') as r: 
            next(r)                   # SKIP HEADERS
            rr = csv.reader(r)
            for row in rr:
                wf.writerow(row)
```
